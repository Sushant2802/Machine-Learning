```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

## 1. Categorical Data

### a. Countplot

```python
sns.countplot(df, x ='Survived', hue='Sex')
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(df, x ='Pclass', hue='Pclass')
```

## b. PieChart

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%.2f')
```

## 2. Numerical Data

### a. Histogram

```python
import matplotlib.pyplot as plt
```

```python
# to check age group majority
plt.hist(df['Age'],bins=9)
```

### b. Distplot

```python
sns.distplot(df['Age'])
```

### c. Boxplot

```python
sns.boxplot(df['Age'], orient='h')
```
