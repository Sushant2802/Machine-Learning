```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('placement1.csv')
```

```python
df.shape
```

```python
df.sample(5)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(16,5))

plt.subplot(121)
sns.distplot(df['cgpa'])

plt.subplot(122)
sns.distplot(df['placement_exam_marks'])

plt.show()
```

```python
# normally distrubuted
df['cgpa'].skew()
```

```python
# right skewed data
df['placement_exam_marks'].skew()
```

```python
df['placement_exam_marks'].describe()
```

```python
# to visualize the outliers for better understanding
sns.boxplot(df['placement_exam_marks'], orient='h')
```

```python
# finding the IQR
percentile25 = df['placement_exam_marks'].quantile(0.25)
print("percentile25 : ", percentile25)

percentile75 = df['placement_exam_marks'].quantile(0.75)
print("percentile25 : ",percentile75)
```

```python
iqr = percentile75 - percentile25
print("iqr : ", iqr)
```

```python
upper_limit = percentile75 + 1.5*iqr
print("upper limit : ", upper_limit)

lower_limit = percentile25 - 1.5*iqr
print("lower limit : ", lower_limit)
```

```python
# the min marks is 0 so we not use here lower_limit
```

```python
df[df['placement_exam_marks'] > upper_limit]
```

```python
df[df['placement_exam_marks'] < lower_limit]
```

### Trimming

```python
new_df = df[df['placement_exam_marks'] < upper_limit]
new_df.shape
```

```python
# comparing

plt.figure(figsize=(20,10))

plt.subplot(221)
sns.distplot(df['placement_exam_marks'])

plt.subplot(222)
sns.boxplot(df['placement_exam_marks'],orient='h' )

plt.subplot(223)
sns.distplot(new_df['placement_exam_marks'])

plt.subplot(224)
sns.boxplot(new_df['placement_exam_marks'], orient='h')

plt.show()
```

### Capping

```python
new_df_cap = df.copy()

# np.where(condition, True, False)
new_df_cap['placement_exam_marks'] = np.where(
                        new_df_cap['placement_exam_marks'] > upper_limit,
                        upper_limit,
                        np.where(
                            new_df_cap['placement_exam_marks'] < lower_limit,
                            lower_limit,
                            new_df_cap['placement_exam_marks']
                        )
)
```

```python
new_df_cap.shape
```

```python
# comparing 

plt.figure(figsize=(16,8))

plt.subplot(221)
sns.distplot(df['placement_exam_marks'])

plt.subplot(222)
sns.boxplot(df['placement_exam_marks'], orient='h')

plt.subplot(223)
sns.distplot(new_df_cap['placement_exam_marks'])

plt.subplot(224)
sns.boxplot(new_df_cap['placement_exam_marks'], orient='h')

plt.show()
```
