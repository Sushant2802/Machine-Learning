```python
import numpy as np
import pandas as pd
```

```python
column_names = ['speed1','speed2','point','rating','size1','size2','pred']
```

```python
df = pd.read_csv('car_evaluation.csv', header=None, names=column_names, usecols=['speed1','size1','pred'])
```

```python
df.sample(9)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('pred',axis=1),
                                                    df['pred'], 
                                                    test_size=0.2,
                                                    random_state=42)
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['low','med','high','vhigh'],['small','med','big']])
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
