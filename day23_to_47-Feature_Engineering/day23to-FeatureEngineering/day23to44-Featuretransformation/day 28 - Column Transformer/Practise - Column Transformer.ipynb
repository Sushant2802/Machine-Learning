```python
from sklearn.base import BaseEstimator, TransformerMixin

class ThresholdTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, threshold=0):
        self.threshold = threshold
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        X = pd.DataFrame(X)
        X = X.applymap(lambda x: 1 if x > self.threshold else 0)
        return X.values
```

```python
# column Transformer
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=
                               [
                                ('trf1', OneHotEncoder(sparse_output=False,  drop='first'),  ['owner','fuel']),
                                ('trf2',  StandardScaler(),  ['km_driven']),
                                ('trf3', ThresholdTransformer(threshold=120), ['brand'])
                               ],
                                 remainder='passthrough')
```

```python
transformer
```

```python
from sklearn.pipeline import Pipeline
pipeline = Pipeline(steps=[
    ('transformer', transformer)
])
```

```python
processed_data = pipeline.fit_transform(df)
print(processed_data)
```
